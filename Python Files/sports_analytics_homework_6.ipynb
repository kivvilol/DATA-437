```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

mlb = pd.read_csv('Teams.csv')
mlb.head()
```

```python
# creating run differential
mlb['RD'] = mlb['R'] - mlb['RA']

# creating win percentage
mlb['Wpct'] = mlb['W'] / (mlb['W'] + mlb['L'])
```

```python
# aggregating data
mlb_agg = mlb[['yearID', 'RD', 'Wpct']]
mlb_agg.head()
```

```python
# selecting when yearID >= 2001
mlb_agg = mlb_agg[mlb_agg['yearID'] >= 2001]
mlb_agg.head()
```

```python
# building a simple linear regression model
mlb_ols = smf.ols(formula = 'Wpct ~ RD', data = mlb_agg).fit()

# extracting results
mlb_ols.summary()
```

```python
# new observation data
newdata = pd.DataFrame({'RD' : [87]})

# predicting salary 
mlb_ols.predict(newdata)

## The win percentage when the run differential is equal to 87 is 55%. 

## Our regression model from part (d) is equal to Y = 0.5 + 0.0006x1. Because our coefficients are greater than 2*std error for both variables
## and the overall model, we can determine that the slope of our line is statistically significant.
```
