```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

teams = pd.read_csv('Teams.csv')
teams.head()
```

```python
my_teams = teams[['yearID', 'lgID', 'teamID', 'G', 'W', 'L', 'R', 'RA']]
my_teams.head()
```

```python
# creating new variables (RD and Wpct)
my_teams['RD'] = my_teams['R'] - my_teams['RA']
my_teams['Wpct'] = my_teams['W'] / (my_teams['W'] + my_teams['L'])
```

```python
# overall linear model
my_teams_ols = smf.ols(formula = 'Wpct ~ RD', data = my_teams).fit()
my_teams_ols.summary()

# predicting Wpct when RD is 10
newdata_0 = pd.DataFrame({'RD' : [10]})
overall_preds = my_teams_ols.predict(newdata_0)
```

```python
# creating decade between 1961 and 1970
mask1 = (my_teams['yearID'] >= 1961) & (my_teams['yearID'] <= 1970)
my_teams1 = my_teams.loc[mask1]

# model for years between 1961 and 1970
my_teams1_ols = smf.ols(formula = 'Wpct ~ RD', data = my_teams1).fit()

# predicting Wpct when RD is 10
preds_1961_1970 = my_teams1_ols.predict(newdata_0)
```

```python
# creating years
mask2 = (my_teams['yearID'] >= 1971) & (my_teams['yearID'] <= 1980)
my_teams2 = my_teams.loc[mask2]

# model for years between 1961 and 1970
my_teams2_ols = smf.ols(formula = 'Wpct ~ RD', data = my_teams2).fit()
my_teams2_ols.summary()

# predicting Wpct when RD is 10
preds_1971_1980 = my_teams2_ols.predict(newdata_0)
```

```python
# creating years
mask3 = (my_teams['yearID'] >= 1981) & (my_teams['yearID'] <= 1990)
my_teams3 = my_teams.loc[mask3]

# model for years between 1961 and 1970
my_teams3_ols = smf.ols(formula = 'Wpct ~ RD', data = my_teams3).fit()
my_teams3_ols.summary()

# predicting Wpct when RD is 10
preds_1981_1990 = my_teams3_ols.predict(newdata_0)
```

```python
# creating years
mask4 = (my_teams['yearID'] >= 1991) & (my_teams['yearID'] <= 2000)
my_teams4 = my_teams.loc[mask4]

# model for years between 1961 and 1970
my_teams4_ols = smf.ols(formula = 'Wpct ~ RD', data = my_teams4).fit()
my_teams4_ols.summary()

# predicting Wpct when RD is 10
preds_1991_2000 = my_teams4_ols.predict(newdata_0)
```

```python
# making results more readable for print statements
result1 = [float(i) for i in overall_preds]
result2 = [float(i) for i in preds_1961_1970]
result3 = [float(i) for i in preds_1971_1980]
result4 = [float(i) for i in preds_1981_1990]
result5 = [float(i) for i in preds_1991_2000]
```

```python
print('The predicted win percentage for a team with a run differential of 10 is', round(result1[0], 5))
print('The predicted win percentage for a team with a run differential of 10 between 1961 and 1970 is', round(result2[0], 5))
print('The predicted win percentage for a team with a run differential of 10 between 1971 and 1980 is', round(result3[0], 5))
print('The predicted win percentage for a team with a run differential of 10 between 1981 and 1990 is', round(result4[0], 5))
print('The predicted win percentage for a team with a run differential of 10 between 1991 and 2000 is', round(result5[0], 5))

## The decade with the highest win percentage is between 1961 and 1970. 
## Also, all of our individual decades have higher win percentages based on 10 run differential than our overall model.
```
